```python
# Write your code here
import pandas as pd
import matplotlib.pyplot as plt
path="/content/drive/MyDrive/netflix_titles.csv"
df=pd.read_csv(path)
df
```

```python
#to get the data of those rows whose gender is null
missing_values=df.isnull().sum()
print(missing_values)


```

```python
# droping all the nan values of director colun and replacing all the nan values of country colun
df.dropna(subset=["director"],inplace=True)
df["country"]=df["country"].fillna("unknown")
df
```

```python
#to delete the coluns show id and descriptions(#once it is drop if you run again it will show error as not found)
df.drop_duplicates(inplace=True)
df.drop(columns=["show_id","description"],axis=1,inplace=True)
df
```

```python
show_types = df['type'].value_counts()
plt.bar(show_types.index, show_types.values)
plt.xlabel('Show Type')
plt.ylabel('Count')
plt.title('Number of Movies vs TV Shows')
plt.show()

```

```python
df['year_added'] = pd.to_datetime(df['date_added']).dt.year
new_titles_by_year = df['year_added'].value_counts().sort_index()
plt.plot(new_titles_by_year.index, new_titles_by_year.values)
plt.xlabel('Year')
plt.ylabel('Number of New Titles')
plt.title('Trend in Number of New Titles Added Each Year')
plt.show()
```

```python
genres = df['listed_in'].str.split(', ', expand=True).stack().value_counts().head(10)
plt.bar(genres.index, genres.values)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Top 10 Most Popular Genres on Netflix')
plt.xticks(rotation=45)
plt.show()
```

```python
df['rating'].value_counts().plot(kind='hist')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Ratings on Netflix')
plt.show()
```

```python
#managing  outliers
# Example: Identifying outliers in a numerical column (e.g., 'Temperature')

# Calculate the mean and standard deviation of the column
mean_temp = df['Temperature(F)'].mean()
std_temp = df['Temperature(F)'].std()

# Define the threshold for outliers (e.g., values outside 3 standard deviations)
lower_threshold = mean_temp - (3 * std_temp)
upper_threshold = mean_temp + (3 * std_temp)

# Identify outliers
outliers = df[(df['Temperature(F)'] < lower_threshold) | (df['Temperature(F)'] > upper_threshold)]

# Handle outliers (e.g., remove them from the DataFrame)
df = df.drop(outliers.index)

# Example: Handling missing values

# Count the number of missing values in each column
missing_values = df.isnull().sum()

# Handle missing values based on the specific column and domain knowledge
# For example, you can drop rows with missing values or impute missing values with mean/median/mode

# Drop rows with missing values
df = df.dropna()

# Impute missing values with mean (replace NaN values with the column mean)
df['Temperature(F)'] = df['Temperature(F)'].fillna(df['Temperature(F)'].mean())

# Impute missing values with median (replace NaN values with the column median)
df['Humidity(%)'] = df['Humidity(%)'].fillna(df['Humidity(%)'].median())

# Impute missing values with mode (replace NaN values with the column mode)
df['Wind_Speed(mph)'] = df['Wind_Speed(mph)'].fillna(df['Wind_Speed(mph)'].mode().iloc[0])
```

```python
# Write your code here
import seaborn as sns
import pandas as pd

# Load the Titanic dataset
titanic_df = sns.load_dataset("titanic")
# Identify missing values in the "age" column
missing_age = titanic_df["age"].isnull().sum()
print("Missing values in the 'age' column:", missing_age)
# Fill missing values in the "age" column with the median age
titanic_df["age"].fillna(titanic_df["age"].median(), inplace=True)
# Convert the data type of the "age" column to numeric(if necessary)
titanic_df["age"] = titanic_df["age"].astype(float)#not necessary here

# Check inconsistent formats in the "sex" column
gender_counts = titanic_df["sex"].value_counts()
print("Unique values in the 'sex' column:\n", gender_counts)
import matplotlib.pyplot as plt

# Box plot of age by survival status
sns.boxplot(x="survived", y="age", data=titanic_df)
plt.title("Box Plot - Age by Survival Status")
plt.xlabel("Survival Status")
plt.ylabel("Age")
plt.show()

# Count plot of gender by survival status
sns.countplot(x="sex", hue="survived", data=titanic_df)
plt.title("Count Plot - Gender by Survival Status")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

```

```python
# Bar plot of survival status by gender
sns.barplot(x="sex", y="survived", data=titanic_df, ci=None)
plt.title("Bar Plot - Survival Status by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Probability")
plt.show()

# Heatmap of age, gender, and survival status
heatmap_data = titanic_df.pivot_table(values="survived", index="age", columns="sex")
sns.heatmap(heatmap_data, cmap="YlGnBu")
plt.title("Heatmap - Survival Probability by Age and Gender")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()
```

```python
# Write your code here
import seaborn as sns
df = sns.load_dataset("iris")
df
```

```python
# Display the first few rows of the dataset
print(df.head())

# Get information about the dataset
print(df.info())

# Generate descriptive statistics of the dataset
print(df.describe())
import matplotlib.pyplot as plt

# Scatter plot
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=df)
plt.title("Sepal Length vs Sepal Width")
plt.show()

# Box plot
sns.boxplot(x="species", y="petal_length", data=df)
plt.title("Petal Length by Species")
plt.show()

# Violin plot
sns.violinplot(x="species", y="petal_width", data=df)
plt.title("Petal Width by Species")
plt.show()

# Pair plot
sns.pairplot(df, hue="species")
plt.title("Pairwise Relationships")
plt.show()

# Heatmap
correlation = df.corr()
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

```
